```python
data.isnull().sum()
```

```python
price_medv = np.array(price_medv).reshape(-1,1)
p_status = np.array(p_status).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    price_medv,
    p_status,
    test_size = 0.2,
    random_state = 3
)
```

Обучим модель

```python
model.fit(X_train, y_train)
```

Посмотрим предсказания модели на тестируемых данных

```python
predicted_y = model.predict(X_test)
```

Посчитаем ошибку модели

**Регрессионный анализ** - можно ли провести линейную зависимость между двумя выборками. Возьмем Ladder - удовлетворение жизнью, и generosity - доброта. Мы хотим найти прямую, которая разделит точки в пространстве наполовину. Это можно сделать графически. Нарисовать график. Можно ли сказать, что данная прямая является собственным вектором? Сделать график регрессионной прямой на данных и собственных векторов. 

```python
ladder = data_2['Ladder']
generosity = data_2['Generosity']
```

```python
ladder = np.array(ladder).reshape(-1,1)
generosity = np.array(generosity).reshape(-1,1)
```

```python
plt.xlabel('Ladder')
plt.ylabel('Generosity')
plt.title('Линейная регрессия от одной переменной')
# Построим прямую для регрессии
# Создадим 100 точек для предсказаний
x = np.linspace(-50, 200, 100)
# Посчитаем предсказанные значения
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(ladder, generosity, 'r.')
plt.xlim()
plt.ylim()
plt.show()
```

**Данная прямая не является собственным вектором**

Странно. Но по точкам видно, что чем больше удовлетворенность жизнью тем больше доброта

```python
import numpy.linalg as la
import math
from sklearn.preprocessing import StandardScaler

ladder = data_2['Ladder']
generosity = data_2['Generosity']

Y = np.vstack((ladder, generosity)).T
w,v = la.eig(np.cov(Y.T))

for i in range(len(w)):
    fig = plt.gcf()
    fig.set_size_inches(6,6)
    plt.suptitle('Собственный вектор № '+ str(i+1), fontsize=16)    
    plt.plot(ladder, generosity, 'g.', zorder=1)
    print('Собственное значение: ', round(w[i],2), 'Собственный вектор: ', v[:,i])
    plt.xlabel('Ladder')
#    plt.xlim(0,7)
    plt.ylabel('Generosity')
#    plt.ylim(-2,5)
    plt.plot([ladder.mean(), generosity.mean() + math.sqrt(w[i]) * v[:,i][0]], 
             [ladder.mean(), generosity.mean() + math.sqrt(w[i]) * v[:,i][1]],color='red', linewidth=3)
    plt.show()
```

моделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
n_experiments = 1000
trials = 100

outcomes = np.empty(n_experiments, dtype=np.float)

for i in range(n_experiments):
    
    values = np.random.randint(1, 7, size=trials)
    
    head_counts = np.sum(values)
    outcomes[i] = head_counts

print('1 experiment with cube: ', values)
plt.hist(outcomes)
plt.xlabel('Количество выпадений значений кубика')
plt.ylabel('Количество встречаемости значения')
plt.show()
```

```python
Почему график изменился?
Из-за достаточно большого количества величин.

Почему можно применить центральную предельную теорему?
Равномерно малое влияние слагаемых на рассеивание суммы
```
